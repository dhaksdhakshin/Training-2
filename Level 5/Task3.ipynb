```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\madhu\\OneDrive\\Documents\\titanic1.csv")
df.head()
```

```python
df[["Age","Fare"]].describe()
```

```python
survival_rates = df.groupby("Pclass")["Survived"].mean()
print(f"Survival Rates:{survival_rates}")
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(8,5))
sns.barplot(x=survival_rates.index, y=survival_rates.values, hue=survival_rates.index , palette="coolwarm")
plt.xlabel("Passanger Class")
plt.ylabel("Survival Rate")
plt.title("Survival Rate by passanger Class")
plt.ylim(0,1)
plt.show()
```
