```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('shopping_trends_updated.csv')
df.head()
```

```python
mean_purchase=df['Purchase Amount (USD)'].mean()
median_purchase=df['Purchase Amount (USD)'].median()
mode_purchase=df['Purchase Amount (USD)'].mode()[0]
```

```python
print(f'Mean Purchase Amount (USD): ${mean_purchase:.2f}')
print(f'median Purchase Amount (USD): ${median_purchase:.2f}')
print(f'mode Purchase Amount (USD): ${mode_purchase:.2f}')
```

```python
def categorize_spenders(amount):
    if amount < mean_purchase :
        return 'Low Spender'
    else: 
        return 'High Spender'
    
df['Spendercategory']=df['Purchase Amount (USD)'].apply(categorize_spenders)
```

```python
df.head()
```

```python
print (df['Spendercategory'].value_counts())
df.head()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x="Spendercategory", palette="coolwarm")
plt.title("Customer Segmentation Based on Purchase Amount")
plt.xlabel("Spender Category")
plt.ylabel("Number of Customers")
plt.show()
```
