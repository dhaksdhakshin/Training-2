```python
import pandas as pd 
import numpy as num 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df=pd.read_csv('shopping_trends_updated.csv')
```

```python
df.head()
```

```python
df['Purchase Amount (USD)'] = pd.to_numeric(df['Purchase Amount (USD)'], errors = 'coerce')
```

```python
df = df.dropna(subset=['Purchase Amount (USD)' ])
```

```python
df.head()
```

```python
df.rename(columns={'Purchase Amount (USD)':'Price'},inplace=True)
```

```python
df.head()
```

```python
price_range = df['Price'].max() - df['Price'].min()
price_variance = df['Price'].var()
price_std_dev = df['Price'].std()
```

```python
print(f'Price Range : {price_range :.2f}')
print(f'Price Variance : {price_variance :.2f}')
print(f'Price Standard Deviation : {price_std_dev :.2f}')
```

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Price"], color="green")
plt.title("Boxplot of Product Prices")
plt.xlabel("Price")
plt.show()
```
